RELIANCE STOCK MARKET PREDICTION

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/ORIGINAL RELIANCE DATASET - ORIGINAL RELIANCE DATASET.csv")
df
```

```python
import math
median_Last = math.floor(df.Last.median())
median_Last
```

```python
df.Last = df.Last.fillna(median_Last)
df


```

```python
import math
median_Trades = math.floor(df.Trades.median())
median_Trades
```

```python
df.Trades = df.Trades.fillna(median_Trades)
df
```

```python
import math
median_Deliverable_Volume = math.floor(df.Deliverable_Volume.median())
median_Deliverable_Volume
```

```python
df.Deliverable_Volume = df.Deliverable_Volume.fillna(median_Deliverable_Volume)
df
```

```python
import math
mean_Percentage_Deliverble = math.floor(df.Percentage_Deliverble.mean())
mean_Percentage_Deliverble
```

```python
df.Percentage_Deliverble = df.Percentage_Deliverble.fillna(mean_Percentage_Deliverble)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Prev_Close','Open','High','Low','Last','Close','VWAP','Volume','Trades','Deliverable_Volume']],df.Turnover)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2633,2688,2756,2675,2370,2896,2729,15337652,354885,16489339]])
```
