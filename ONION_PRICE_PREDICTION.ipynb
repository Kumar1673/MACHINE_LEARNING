ONION PRICE PREDICTION

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("Onion Prices 2020.csv")
df
```

```python
import math
median_max_price = math.floor(df.max_price.median())
median_max_price
```

```python
df.max_price= df.max_price.fillna(median_max_price)
df
```

```python
import math
median_modal_price = math.floor(df.modal_price.median())
median_modal_price
```

```python
df.modal_price= df.modal_price.fillna(median_modal_price)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['min_price','max_price']],df.modal_price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[4000,500]])
```
